```python
def parse_line(line):
    constraint, password = line.strip().split(': ')
    pos, letter = constraint.split(' ')
    pos1, pos2 = pos.split('-')
    
    return int(pos1)-1, int(pos2)-1, letter, password
```

```python
with open('input.txt', 'r') as f:
    df = pd.DataFrame([parse_line(l) for l in f.readlines()], columns=['idx1', 'idx2', 'letter', 'password'])
    
df
```

```python
def get_letters(idx, password):
    return list(password)[idx]
```

```python
df.loc[:, 'p1'] = df.apply(lambda row: get_letters(row['idx1'], row['password']), axis=1)
df.loc[:, 'p2'] = df.apply(lambda row: get_letters(row['idx2'], row['password']), axis=1)

df
```

```python
mask = (df['p1'] == df['letter']) ^ (df['p2'] == df['letter'])
mask.value_counts()
```

```python
df.loc[mask[mask].sample(5).index.tolist(), :]
```

```python
df.loc[mask[~mask].sample(5).index.tolist(), :]
```
