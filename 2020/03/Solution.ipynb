# Part 1

```python
def count_trees(drow, dcol):
    (irow, icol) = (0, 0)

    num_trees = 0
    while True:
        irow += drow
        if irow >= num_rows:
            break
    
        icol += dcol
        icol = icol % num_cols
    
        # print(irow, icol, arr[irow, icol])
    
        if arr[irow, icol] == '#':
            num_trees += 1
        
    return num_trees
```
